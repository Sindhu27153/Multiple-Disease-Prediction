```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
liver_df = pd.read_csv("indian_liver_patient - indian_liver_patient.csv")
```

```python
liver_df
```

```python
liver_df.columns = liver_df.columns.str.lower()
```

```python
liver_df["albumin_and_globulin_ratio"] = liver_df["albumin_and_globulin_ratio"].fillna(liver_df["albumin_and_globulin_ratio"].median())
```

```python
# Check status balance
liver_df['dataset'].value_counts().plot(kind='barh').set_title('counts for liver Disease(1 = Present, 2 = Absent)')
```

```python
num_cols = liver_df.select_dtypes(include=['int64','float64']).columns
num_cols = num_cols.difference(['dataset'])

liver_df[num_cols].hist(figsize=(12, 12), bins=20, color='purple', edgecolor='black')  
plt.suptitle("Distribution of Numerical Features")  
plt.show() 
```

```python
#gender vs dataset
plt.figure(figsize=(12,6))
sns.barplot(data=liver_df,x = 'gender',y='dataset',palette='coolwarm')
plt.show()
```

```python
#Heatmap to find correlation Between values
corr_matrix = liver_df.corr(numeric_only=True)

plt.figure(figsize=(12,6))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
#Numercical columns vs Dataset
import math
n_cols = 3  # Number of columns in the subplot grid
n_rows = math.ceil(len(num_cols) / n_cols)  # Automatically determine the required rows

plt.figure(figsize=(15, 5 * n_rows))  # Adjust figure size based on rows

for i, col in enumerate(num_cols, 1):  
    plt.subplot(n_rows, n_cols, i)
    sns.scatterplot(data=liver_df, x="dataset", y=col, hue="dataset", palette="coolwarm")  
    plt.title(col)  
plt.tight_layout()  
plt.title(f"Distribution of {col} vs dataset")
plt.show()
```

```python
plt.figure(figsize=(24,12))
sns.boxplot(liver_df)
plt.title("Distribution of numerical values")
plt.show()
```

```python


# Apply IQR capping
for col in num_cols:
    Q1 = liver_df[col].quantile(0.25)
    Q3 = liver_df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Capping outliers
    liver_df[col] = liver_df[col].clip(lower_bound, upper_bound)

print("Outliers capped using IQR method.")
```

```python
#numerical columns to standardized column values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

for col in num_cols:
    liver_df[col] = scaler.fit_transform(liver_df[[col]])

liver_df.head(5)
```

```python
#one hot encoding for gender columns
liver_df = pd.get_dummies(liver_df, columns=['gender'], drop_first=True)
liver_df
```

```python
liver_df["gender_Male"] = liver_df["gender_Male"].astype(int)
liver_df
```

```python
#Feature Scaling
x = liver_df.drop(columns=['dataset'])
y = liver_df['dataset']
```

```python
#Train test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import precision_score, recall_score, f1_score,roc_auc_score

liver_model = LogisticRegression()
liver_model.fit(x_train,y_train)
y_pred = liver_model.predict(x_test)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
auc = roc_auc_score(y_test,y_pred)
print(f"precision score: {precision}")
print(f"recall score:{recall}")
print(f"f1 score:{f1}")
print(f"roc_auc score:{auc}")
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, recall_score, f1_score,roc_auc_score

liver_model = DecisionTreeClassifier(max_depth=4)
liver_model.fit(x_train,y_train)
y_pred = liver_model.predict(x_test)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
auc = roc_auc_score(y_test,y_pred)
print(f"precision score: {precision}")
print(f"recall score:{recall}")
print(f"f1 score:{f1}")
print(f"roc_auc score:{auc}")
```

```python
#Random Forest classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, f1_score,roc_auc_score

liver_model = RandomForestClassifier(n_estimators=100,random_state=42)
liver_model.fit(x_train,y_train)
y_pred = liver_model.predict(x_test)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
auc = roc_auc_score(y_test,y_pred)
print(f"precision score: {precision}")
print(f"recall score:{recall}")
print(f"f1 score:{f1}")
print(f"roc_auc score:{auc}")
```

```python
#K Nearest Neignbors classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_score, recall_score, f1_score

liver_model = KNeighborsClassifier()
liver_model.fit(x_train,y_train)
y_pred = liver_model.predict(x_test)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
auc = roc_auc_score(y_test,y_pred)
print(f"precision score: {precision}")
print(f"recall score:{recall}")
print(f"f1 score:{f1}")
print(f"roc_auc score:{auc}")
```
